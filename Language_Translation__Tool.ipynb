```python
import googletrans
from googletrans import Translator, LANGUAGES
import ipywidgets as widgets
from IPython.display import display, HTML

# Initialize the translator
translator = Translator()

# Create the widgets for input text, output text, and language selection
input_text = widgets.Textarea(
    value='',
    placeholder='Enter text to translate...',
    description='',
    layout=widgets.Layout(width='90%', height='100px')
)

output_text = widgets.Textarea(
    value='',
    placeholder='Translated text will appear here...',
    description='',
    disabled=True,
    layout=widgets.Layout(width='90%', height='150px')
)

language_dropdown = widgets.Dropdown(
    options=[(language.title(), code) for code, language in LANGUAGES.items()],
    value='en',  # Default to English
    description='',
    layout=widgets.Layout(width='50%')
)

translate_button = widgets.Button(
    description="Translate",
    button_style='success',
    tooltip="Click to translate",
    icon='language',
    layout=widgets.Layout(width='20%')
)

# Translate function
def on_translate_button_clicked(b):
    text = input_text.value
    dest_lang = language_dropdown.value
    if text.strip():
        try:
            translated = translator.translate(text, dest=dest_lang)
            output_text.value = translated.text
        except Exception as e:
            output_text.value = f"Error: {str(e)}"
    else:
        output_text.value = "Please enter some text to translate."

# Link the translate function to the button
translate_button.on_click(on_translate_button_clicked)

# HTML Styling
style = """
<style>
    .custom-title {
        font-size: 24px;
        font-weight: bold;
        color: #f9f9f9 ;
        text-align: center;
        margin-bottom: 20px;
    }
    .custom-container {
        border: 1px solid #ccc;
        border-radius: 10px;
        padding: 20px;
        background-color: #4B0082 ;
        width: 80%;
        margin: auto;
        box-shadow: 0 0 10px rgba(0,0,0,0.1);
    }
    .custom-label {
        font-size: 18px;
        font-weight: bold;
        margin-bottom: 10px;
        color: #f9f9f9;
    }
</style>
"""

# Create an HTML container for styling
html_header = HTML(f"""
<div class="custom-container">
    <div class="custom-title">Language Translator Tool</div>
    <div class="custom-label">Input Text</div>
</div>
""")

html_language_label = HTML(f"""
<div class="custom-label">Select Target Language</div>
""")

html_output_label = HTML(f"""
<div class="custom-label">Translated Text</div>
""")

# Display styled UI components
display(HTML(style))  # Apply CSS
display(html_header)
display(input_text)
display(html_language_label)
display(language_dropdown, translate_button)
display(html_output_label)
display(output_text)
```
